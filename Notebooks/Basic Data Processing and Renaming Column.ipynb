# Translation Service

```python
from googletrans import Translator, constants
```

```python
def Translation_Service(string_var, dest = 'en', src = 'ko'):
    translated_var = Translator().translate(string_var, dest=dest, src=src)
    return translated_var.text
```

## Cancer Types

```python
cancer_types.columns
```

```python
cancer_types['Classification_en'] = cancer_types['분류'].apply(Translation_Service)
cancer_types['Contents_en']=cancer_types['내용'].apply(Translation_Service)
```

### DODO Translation Content

```python
Path = 'C:/Users/Tanz/Desktop/Dodo/ReDiscovery_Cancer_BigData/Notebooks/1. 유방암 Fake Data/'
```

```python
Files = os.listdir(Path)
Files
```

```python
OUTPUT_PATH = ''
i_en = []
for i in Files:
    print(i)
    file_name = i.split('.')[0]
    data = pd.read_csv(Path+i, encoding='CP949')
    i_en_name = Translation_Service(file_name)
    i_en.append(i_en_name)
    en_column = []
    for j in data.columns:
        if Translator().detect(j).lang == 'ko':
            en_column.append(Translation_Service(j))
        else:
            en_column.append(j)
            
    data.columns = en_column
    
    data.to_csv('C:/Users/Tanz/Desktop/Dodo/ReDiscovery_Cancer_BigData/Notebooks/1. Breast_Cancer Fake Data - EN/'+i_en_name+'.csv', encoding = 'utf_8_sig', index=False   )
```
